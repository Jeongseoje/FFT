```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft

# Generate a sample time series data
t = np.linspace(0, 1, 400, endpoint=False)
# signal = np.sin(2 * np.pi * 5 * t) + np.sin(2 * np.pi * 10 * t)  # A sum of two sine waves
signal = np.sin(1 * np.pi * 25 * t) # A one sine waves

# Apply Fourier transform
freq_domain = fft(signal)
```

```python
freq_domain
```

```python
# Compute the frequency bins
N = len(t)
T = t[1] - t[0]  # sample spacing
f = np.linspace(0.0, 1.0/(2.0*T), N//2)

# Compute the magnitude of the Fourier coefficients
magnitude = 2.0/N * np.abs(freq_domain[:N//2])

# Plot the spectrum
plt.figure(figsize=(6, 4))
plt.plot(f, magnitude)
plt.title('Frequency Spectrum of the Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python

# Apply inverse Fourier transform
time_series_reconstructed = ifft(freq_domain)

# Plot the original and reconstructed signals
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, signal)
plt.title('Original Signal')

plt.subplot(2, 1, 2)
plt.plot(t, time_series_reconstructed.real)  # Taking the real part as the output is complex
plt.title('Reconstructed Signal from Fourier Transform')

plt.tight_layout()
plt.show()
```

```python

# Generate a sample time series data
t = np.linspace(0, 1, 400, endpoint=False)
signal = np.sin(2 * np.pi * 5 * t) + np.sin(2 * np.pi * 10 * t)  # A sum of two sine waves

# Apply Fourier transform
freq_domain = fft(signal)
```

```python
# Compute the frequency bins
N = len(t)
T = t[1] - t[0]  # sample spacing
f = np.linspace(0.0, 1.0/(2.0*T), N//2)

# Compute the magnitude of the Fourier coefficients
magnitude = 2.0/N * np.abs(freq_domain[:N//2])

# Plot the spectrum
plt.figure(figsize=(6, 4))
plt.plot(f, magnitude)
plt.title('Frequency Spectrum of the Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
np.max(magnitude)
```

```python

# Apply inverse Fourier transform
time_series_reconstructed = ifft(freq_domain)

# Plot the original and reconstructed signals
plt.figure(figsize=(12, 12))

plt.subplot(3, 1, 1)
plt.plot(t, signal)
plt.title('Original Signal')

plt.subplot(3, 1, 2)
plt.plot(t, time_series_reconstructed.real)  # Taking the real part as the output is complex
plt.title('Reconstructed Signal from Fourier Transform')

plt.subplot(3, 1, 3)
plt.plot(f, magnitude)
plt.title('Frequency Spectrum of the Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()

plt.tight_layout()
plt.show()
```
